```
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```
def sin_data():
    x = np.linspace(0, 2 * np.pi, 1000)
    y = np.sin(x)
    return (x, y)


def cos_data():
    x = np.linspace(0, 2 * np.pi, 1000)
    y = np.cos(x)
    return (x, y)
```

```
x, y = sin_data()
plt.plot(x, y)
```

```
plt.style.available
```

```
plt.style.use('classic')
```

```
x_sin, y_sin = sin_data()
x_cos, y_cos = cos_data()
plt.grid()
plt.plot(x_sin, y_sin)
plt.plot(x_cos, y_cos, 'g*')
```

```
# created empty figure
fig = plt.figure()

# adding axes to the figure
ax1 = fig.add_axes((0, 0, 2, 2))

ax2 = fig.add_axes((1.4, 1.4, 0.6, 0.6))
      

ax1.plot(x_sin, y_sin)
ax2.plot(x_cos, y_cos, 'r')
```

```
from matplotlib import animation
```

```
help(animation.FuncAnimation)
```

```
fig = plt.figure()
ax1 = fig.add_axes((0, 0, 1, 1), xlim=(-0.1, 2 * np.pi), ylim=(-1.1, 1.1))
img, = ax1.plot([], [])
```

```
def init():
    img.set_data([], [])
    return img,

def animate(i):
    x = np.linspace(0  * 10, 2.5 * np.pi, 1000)
    added_stuff = np.full((1, 1000), 1, dtype=float)
    y = np.sin(x + i * added_stuff)
    img.set_data(x, y)
    return img,
```

```
movie = animation.FuncAnimation(fig, animate, init_func=init, frames=300, interval=20, blit=True)
movie.save("my_movie.mp4", fps=15, extra_args=['-vcodec', 'libx264'])
```
